```python
# load packages
from imutils import face_utils
import imutils
import dlib
import cv2
import glob
import numpy as np
from matplotlib import pyplot as plt
from face_functions import *
```

```python
# load the images
#image1_name = 'files/mona_lisa.jpg'
#image2_name = 'files/headshot.jpg'

image2_name = 'files/mona_lisa.jpg'
image1_name = 'files/headshot.jpg'
```

```python
# compute the face cropped file for each image

# for the art image
image1 = cv2.imread(image1_name,cv2.IMREAD_COLOR)
image1_faces, image1_face_detect, image1_cropped_image, image1_rectangle = compute_face_crop(image1)
plt.figure()
plt.imshow(image1_cropped_image)

# for the headshot image
image2  = cv2.imread(image2_name,cv2.IMREAD_COLOR)
image2_faces, image2_face_detect, image2_cropped_image, image2_rectangle = compute_face_crop(image2)
plt.figure()
plt.imshow(image2_cropped_image)
```

```python
# compute the face landmarks
# load the predictor from dlib
p = "files/shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
# compute the landmarks in the art and headshot faces
# for the art image
image1_points, image1_points1, image1_cropped_lm_image = compute_landmarks(image1_cropped_image,detector,predictor)
plt.figure()
plt.imshow(image1_cropped_lm_image)

# for the headshot image
image2_points, image2_points1, image2_cropped_lm_image = compute_landmarks(image2_cropped_image,detector,predictor)
plt.figure()
plt.imshow(image2_cropped_lm_image)
```

```python
# find landmarks for the center of the eye
l_eye_image1 = image1_points[42:48]
r_eye_image1 = image1_points[36:42]
r_eye_image1_center = find_center(image1_cropped_image, r_eye_image1)
l_eye_image1_center = find_center(image1_cropped_image, l_eye_image1)
image1_points.append(r_eye_image1_center)
image1_points.append(l_eye_image1_center)

l_eye_image2 = image2_points[42:48]
r_eye_image2 = image2_points[36:42]
r_eye_image2_center = find_center(image2_cropped_image, r_eye_image2)
l_eye_image2_center = find_center(image2_cropped_image, l_eye_image2)
image2_points.append(r_eye_image2_center)
image2_points.append(l_eye_image2_center)

# find the landmark for the center of the mouth
mouth_image1 = image1_points[48:59]
mouth_image1_center = find_center(image1_cropped_image, mouth_image1)
image1_points.append(mouth_image1_center)

mouth_image2 = image2_points[48:59]
mouth_image2_center = find_center(image2_cropped_image, mouth_image2 )
image2_points.append(mouth_image2_center)
```

```python
# compute the convex hull for each image
hullIndex_image1,hullIndex_image1_1 = computeHullIndex(image1_points1, image1_points)
hullIndex_image2,hullIndex_image2_1 = computeHullIndex(image2_points1, image2_points)

# compute the convex hull for one image - corresponds to other
hull_image1 = []
hull_image2 = []
for i in range(0, len(hullIndex_image1_1)):
    hull_image2.append(image2_points[int(hullIndex_image1_1[i])])
    hull_image1.append(image1_points[int(hullIndex_image1_1[i])])
    
# compute the convex hull for one image - corresponds to other
hull_image1_1 = []
hull_image2_1 = []
for i in range(0, len(hullIndex_image2_1)):
    hull_image2_1.append(image2_points[int(hullIndex_image2_1[i])])
    hull_image1_1.append(image1_points[int(hullIndex_image2_1[i])])
```

```python
# decide which landmarks to keep
image1_points_final = image1_points
# art_points_final.append(art_points[36]) # tip of nose
# art_points_final.append(art_points[39]) # tip of nose
# art_points_final.append(art_points[42]) # tip of nose
# art_points_final.append(art_points[45]) # tip of nose
# art_points_final.append(art_points[48]) # l mouth corner
# art_points_final.append(art_points[54]) # r mouth corner
# art_points_final.append(art_points[68]) # r eye center
# art_points_final.append(art_points[69]) # l eye center
# art_points_final.append(art_points[70]) # mouth middle

image2_points_final = image2_points
# headshot_points_final.append(headshot_points[36]) # tip of nose
# headshot_points_final.append(headshot_points[39]) # tip of nose
# headshot_points_final.append(headshot_points[42]) # tip of nose
# headshot_points_final.append(headshot_points[45]) # tip of nose
# headshot_points_final.append(headshot_points[48]) # l mouth corner
# headshot_points_final.append(headshot_points[54]) # r mouth corner
# headshot_points_final.append(headshot_points[68]) # r eye center
# headshot_points_final.append(headshot_points[69]) # l eye center
# headshot_points_final.append(headshot_points[70]) # mouth middle
```

```python
# find the Delaunay Triangulation for each image
dt_image1, image1_triangles = calculateDelaunayTriangles(image1_cropped_image, image1_points_final)
plt.figure()
plt.imshow(image1_triangles)

# find the Delaunay Triangulation for each image
dt_image2, image2_triangles  = calculateDelaunayTriangles(image2_cropped_image, image2_points_final)
plt.figure()
plt.imshow(image2_triangles)
```

```python
# compute the warped version of the headshot image
image2_warped1 = compute_warped_image(image2_cropped_image,image1_cropped_image,dt_image1,image2_points_final,image1_points_final)
plt.imshow(image2_warped1)
```

```python
# warp this face back into the original mona lisa
# mona lisa is currently in the convex hull of headshot - calculate triangles again

# find the Delaunay Triangulation for each image
dt_image2, image2_triangles = calculateDelaunayTriangles(image2_cropped_image, hull_image2)
plt.figure()
plt.imshow(image2_triangles)

# find the Delaunay Triangulation for each image
dt_image1, image1_triangles  = calculateDelaunayTriangles(image2_warped1, hull_image1)
plt.figure()
plt.imshow(image1_triangles)
```

```python
# compute the mask and seamlessly clone it into the art image
image2_warped2 = compute_warped_image(image2_warped1,image2_cropped_image,dt_image1,hull_image1,hull_image2)
new_image, mask = clone_image(hullIndex_image2,image2_warped2,image2_cropped_image)

plt.figure()
plt.imshow(image2_cropped_image)
plt.figure()
plt.imshow(new_image)
plt.figure()
plt.imshow(image1_cropped_image)
```

```python
# align the images

# compute landmarks for the new image
newimage_points, newimage_points1, newimage_cropped_lm_image = compute_landmarks(new_image,detector,predictor)
plt.figure()
plt.imshow(newimage_cropped_lm_image)

# get the shape of the newimage, compute new image2, find landmarks in coord system
new_shape = np.shape(new_image)
image1_cropped_scaled = imutils.resize(image1_cropped_image, width=new_shape[0], height = new_shape[1])
image1_resized_points, image1_resized_points1, image1_resized_cropped_lm_image = compute_landmarks(image1_cropped_scaled,detector,predictor)

point1 = 9
point2 = 28
dest_points = [ newimage_points[point1], newimage_points[point2] ] 
source_points  = [ image1_resized_points[point1], image1_resized_points[point2] ] 
translated_image1 = align_image(image1_cropped_scaled, source_points, dest_points, w = new_shape[0], h = new_shape[1])

image1_resized_points, image1_resized_points1, image1_resized_cropped_lm_image = compute_landmarks(translated_image1,detector,predictor)
point1 = 4
point2 = 14
dest_points = [ newimage_points[point1], newimage_points[point2] ] 
source_points  = [ image1_resized_points[point1], image1_resized_points[point2] ] 
translated_image1 = align_image(translated_image1, source_points, dest_points, w = new_shape[0], h = new_shape[1])

plt.figure()
plt.imshow(new_image)
plt.figure()
plt.imshow(translated_image1)

output_name = 'files/output_image.jpg'
cv2.imwrite('files/art_warped_in_art.jpg',new_image)
cv2.imwrite('files/headshot_aligned.jpg',translated_image1)

#cv2.imwrite('files/headshot_warped.jpg',new_image)
#cv2.imwrite('files/art_aligned.jpg',translated_image1)
```
