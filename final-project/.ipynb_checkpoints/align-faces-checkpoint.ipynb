```python
# load packages
from imutils import face_utils
import imutils
import dlib
import cv2
import glob
import numpy as np
from matplotlib import pyplot as plt
from face_functions import *
```

```python
# load the images
image1_name = 'files/mona_lisa.jpg'
image2_name = 'files/headshot.jpg'

#image2_name = 'files/mona_lisa.jpg'
#image1_name = 'files/headshot.jpg'
```

```python
# compute the face cropped file for each image

# for the art image
image1 = cv2.imread(image1_name,cv2.IMREAD_COLOR)
image1_faces, image1_face_detect, image1_cropped_image, image1_rectangle = compute_face_crop(image1)
plt.figure()
plt.imshow(image1_cropped_image)

# for the headshot image
image2  = cv2.imread(image2_name,cv2.IMREAD_COLOR)
image2_faces, image2_face_detect, image2_cropped_image, image2_rectangle = compute_face_crop(image2)
plt.figure()
plt.imshow(image2_cropped_image)
```

```python
# compute the face landmarks
# load the predictor from dlib
p = "files/shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
# compute the landmarks in the art and headshot faces
# for the art image
image1_points, image1_points1, image1_cropped_lm_image = compute_landmarks(image1_cropped_image,detector,predictor)
plt.figure()
plt.imshow(image1_cropped_lm_image)

# for the headshot image
image2_points, image2_points1, image2_cropped_lm_image = compute_landmarks(image2_cropped_image,detector,predictor)
plt.figure()
plt.imshow(image2_cropped_lm_image)
```

```python
# get the shape of the newimage, compute new image2, find landmarks in coord system
new_shape = np.shape(image2_cropped_image)
image1_cropped_scaled = imutils.resize(image1_cropped_image, width=new_shape[0], height = new_shape[1])
image1_resized_points, image1_resized_points1, image1_resized_cropped_lm_image = compute_landmarks(image1_cropped_scaled,detector,predictor)

point1 = 9
point2 = 28
dest_points = [ image2_points[point1], image2_points[point2] ] 
source_points  = [ image1_resized_points[point1], image1_resized_points[point2] ] 
translated_image1 = align_image(image1_cropped_scaled, source_points, dest_points, w = new_shape[0], h = new_shape[1])

image1_resized_points, image1_resized_points1, image1_resized_cropped_lm_image = compute_landmarks(translated_image1,detector,predictor)
point1 = 4
point2 = 14
dest_points = [ image2_points[point1], image2_points[point2] ] 
source_points  = [ image1_resized_points[point1], image1_resized_points[point2] ] 
translated_image1 = align_image(translated_image1, source_points, dest_points, w = new_shape[0], h = new_shape[1])

plt.figure()
plt.imshow(image2_cropped_image)
plt.figure()
plt.imshow(translated_image1)

output_name = 'files/output_image.jpg'
#cv2.imwrite('files/art_for_translated_hs.jpg',image2_cropped_image)
#cv2.imwrite('files/translated_hs.jpg',translated_image1)

cv2.imwrite('files/translated_art.jpg',image2_cropped_image)
cv2.imwrite('files/hs_for_translate_art.jpg',translated_image1)
```
