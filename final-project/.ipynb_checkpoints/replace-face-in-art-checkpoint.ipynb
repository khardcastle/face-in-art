```python
# load packages
from imutils import face_utils
import dlib
import cv2
import glob
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load the images
art_image = 'files/mona_lisa.jpg'
headshot_image = 'files/headshot.jpg'
```

```python
# crop the image so it's just the face
def compute_face_crop(image_file):
    
    # This will take an image file, and will return information about where the face is, the image with the face bounding box, and the cropped image
    
    # read image, copy it so nothing funky happens, convert to grayscale for opencv
    img = cv2.imread(image_file,cv2.IMREAD_COLOR)
    imgtest = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # load the face classifier
    facecascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    
    # run the face classifier for the image. returns bounding box of the face
    faces = facecascade.detectMultiScale(imgtest, scaleFactor=1.2, minNeighbors=5)
    # plot the image and the bounding box of the face
    if len(faces) > 2:
        display('too many faces')
        faces = []
        face_detect = []
        cropped_image = img
        return
    else:
        (x, y, w, h) = faces[0]
        pad = 25;
        image_size = np.shape(imgtest)
        print(image_size)
        # detect the face region
        minx = max(0,x-pad)
        miny= max(0,y-pad)
        maxx = min(image_size[1],x+w+pad)
        maxy = min(image_size[0],y+w+pad) 
        face_detect = cv2.rectangle(imgtest, (minx, miny), (maxx, maxy), (255, 0, 255), 2)

        # save the cropped image
        cropped_image = img[miny:maxy, minx:maxx,:]
        
        return faces, face_detect, cropped_image
```

```python
# compute the face cropped file for each image

# for the art image
art_faces, art_face_detect, art_cropped_image = compute_face_crop(art_image)
plt.imshow(art_cropped_image)

# for the headshot image
headshot_faces, headshot_face_detect, headshot_cropped_image = compute_face_crop(headshot_image)
plt.imshow(headshot_cropped_image)
```

```python
# compute the face landmarks
# load the predictor from dlib
p = "files/shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
def compute_landmarks(image_file):
    #img = cv2.imread(image_file,cv2.IMREAD_COLOR) # need to change this to be scipy version of reading in images
    img = image_file
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
    detected_image = detector(img, 0)
    shape = predictor(img, detected_image[0])
    shape = face_utils.shape_to_np(shape)
    
    # draw landmarks on face
    for (x, y) in shape:
        cv2.circle(img, (x, y), 2, (0, 0, 0), -1)
    
    return shape, img
```

```python
# compute the landmarks in the art and headshot faces

# for the art image
art_shape, art_cropped_lm_image = compute_landmarks(art_cropped_image)
plt.imshow(art_cropped_lm_image)

# for the headshot image
headshot_shape, headshot_cropped_lm_image = compute_landmarks(headshot_cropped_image)
plt.imshow(headshot_cropped_lm_image)
```

```python
# Check if a point is inside a rectangle
def rectContains(rect, point) :
    if point[0] < rect[0] :
        return False
    elif point[1] < rect[1] :
        return False
    elif point[0] > rect[0] + rect[2] :
        return False
    elif point[1] > rect[1] + rect[3] :
        return False
    return True
```
