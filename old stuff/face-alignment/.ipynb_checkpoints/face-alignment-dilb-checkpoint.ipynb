```python
from imutils import face_utils
import dlib
import cv2
import glob
from matplotlib import pyplot as plt
```

```python
p = "shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
img = cv2.imread("headshot_kiahhardcastle_cropped.jpg",cv2.IMREAD_COLOR) # need to change this to be scipy version of reading in images
imgtest1 = img.copy()
imgtest = cv2.cvtColor(imgtest1, cv2.COLOR_BGR2GRAY)  
detected_image = detector(imgtest, 0)
plt.imshow(imgtest)
```

```python
shape = predictor(imgtest, detected_image)


shape = face_utils.shape_to_np(shape)
    
# Draw on our image, all the finded cordinate points (x,y) 
for (x, y) in shape:
    cv2.circle(image, (x, y), 2, (0, 255, 0), -1)
    
# Show the image
cv2.imshow("Output", image)
```
